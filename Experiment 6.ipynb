## Experiment 6

Code to import Pandas and NumPy library.

```python
import pandas as pd
import numpy as np
```

Program to print first 20 rows.

```python
df = pd.read_csv('AmesHousing.csv')
df.head(20)
```

Program to replace NaN value in Lot Frontage column by their mean value.

```python
df.fillna(df['Lot Frontage'].mean(),inplace=True)
```

Program to print first 20 rows after replacing NaN values in Lot Frontage by Mean.

```python
df.head(20)
```

Program to find 25th and 75th percentile .

```python
Q1 = df['Lot Frontage'].quantile(0.25)
Q3 = df['Lot Frontage'].quantile(0.75)
print(Q1,Q3)
```

Program to find IQR value.

```python
IQR = Q3-Q1
IQR
```

Program to find lower limit and upper limit values.

```python
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
lower_limit,upper_limit
```

Program to find rows which has value lesser than lower_limit and greater than upper_limit.

```python
df[(df['Lot Frontage']<lower_limit) | (df['Lot Frontage']>upper_limit)]
```

Program to find rows after removing rows which has outliers.

```python
df_new = df[(df['Lot Frontage']>lower_limit) & (df['Lot Frontage']<upper_limit)]
df_new.head(20)
```

Program to check if still their exists null value.

```python
print(df_new.isnull().sum())
```

Program to check if still their exists values lesser than lower_limit and greater than upper_limit.

```python
any_outline = df_new[(df_new['Lot Frontage']<lower_limit) | (df_new['Lot Frontage']>upper_limit)]
any_outline
```

Program to find shape of old and new csv file.

```python
print("Original shape:", df.shape)
print("New shape after removing outliers:", df_new.shape)
```

Program to create new csv file.

```python

df_new.to_csv('AmesHousing_cleaned2.csv')
```
